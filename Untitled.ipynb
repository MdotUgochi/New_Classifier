```python
#initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
```

```python
#Preprocessing
def preprocess_text(text):
    text = text.lower()
    tokens = nltk.word_tokenize(text)
    tokens = [t for t in tokens if t not in string.punctuation]
    tokens = [t for t in tokens if t not in stop_words]
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    return " ".join(tokens)

df["clean_text"] = df["text"].apply(preprocess_text)
```

```python
df
```


Preprocessing done. all text in lower case, lemmatized and stopwords removed. Preprocessed text is stored in a new column labeled "clean_text"

```python
#Vectorize
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 1))
X = vectorizer.fit_transform(df["clean_text"])
Y = df["category"]
```

```python
#Confirm vectorizer is done
X.shape
```

Vectorization done and confirmed

```python
#import modules classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.naive_bayes import MultinomialNB
```

```python
#split data
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

#train model with a Naive Bayes model
model = MultinomialNB()
model.fit(X_train, Y_train)

#check predictions
y_pred = model.predict(X_test)
print(classification_report(Y_test, y_pred))
```

```python
#VisuaLization
cm = confusion_matrix(Y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
feature_names = vectorizer.get_feature_names_out()
class_labels = model.classes_

for i, label in enumerate(class_labels):
    top10 = np.argsort(model.feature_log_prob_[i])[-10:]
    print(label, [feature_names[j] for j in top10])
```
